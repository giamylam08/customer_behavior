# **TỔNG QUAN**

Đây là một tập dữ liệu giao dịch hàng hóa trên nhiều quốc gia, chứa tất cả các giao dịch diễn ra từ ngày 01/12/2010 đến 09/12/2011 của một công ty bán lẻ trực tuyến tại Vương quốc Anh. Công ty này chủ yếu bán các quà tặng độc đáo dành cho mọi dịp lễ. Nhiều khách hàng của công ty là các nhà bán buôn.

Nhằm cung cấp cho chủ cửa hàng những cái nhìn tổng quan về hành vi khách hàng của họ, dự án sẽ tập trung vào các mục tiêu phân tích sau
- Tỷ lệ khách hàng quay lại (Retention Rate)
- Phân khúc khách hàng - Phân tích mô hình RFM
- Phân tích giỏ hàng - Basket Analysis.

## **Key takeaways**


1. **Doanh thu từ khách hàng cũ** thể hiện sự tăng trưởng ổn định hơn so với khách hàng mới, điều này nhấn mạnh tầm quan trọng của việc duy trì và chăm sóc khách hàng hiện tại.

2. **Tỷ lệ khách hàng quay lại sau lần mua hàng đầu tiên thấp do đặc thù sản phẩm là mặt hàng không thiết yếu**: với sản phẩm quà tặng không thiết yếu, tỷ lệ giữ chân khách hàng thấp là điều dễ hiểu, đặc biệt khi khách hàng thường nhập hàng theo mùa vụ và còn có những yếu tố khác như tình trạng hàng tồn kho và tình hình kinh doanh của các đại lý cũng ảnh hưởng đến việc họ có tiếp tục quay lại mua hàng hay không. Tỷ lệ quay lại của cửa hàng thấp hơn mức trung bình của ngành bán lẻ, cho thấy cửa hàng đang gặp khó khăn trong việc khuyến khích khách hàng quay lại mua sắm.

3. **Sự giảm dần của khách hàng sau lần giao dịch đầu tiên**: Việc giảm dần số lượng khách hàng mới và khách quay lại sau giao dịch đầu tiên là tín hiệu cảnh báo.

4. **Nhóm khách hàng có tần suất mua thấp và có nguy cơ rời bỏ** cần được chú ý và khuyến khích quay lại thông qua các chiến dịch tiếp thị phù hợp.

5. Cần triển khai **các chiến dịch marketing** hiệu quả hơn để cải thiện tỷ lệ giữ chân khách hàng và thu hút khách hàng mới, đặc biệt trong những thời điểm doanh thu suy giảm.

## **Gợi ý**

1. **Duy trì và chăm sóc khách hàng cũ**:
   - **Phát triển chương trình khách hàng thân thiết**: Tạo ra các ưu đãi, điểm thưởng và chương trình đặc biệt để khuyến khích khách hàng cũ quay lại mua sắm.
   - **Cá nhân hóa trải nghiệm mua sắm**: Sử dụng dữ liệu lịch sử mua hàng để gửi các đề xuất sản phẩm phù hợp với sở thích của khách hàng.

2. **Tăng cường tỷ lệ quay lại**:
   - **Chạy các chương trình khuyến mãi trong tháng thứ hai**: Nhắm vào những khách hàng đã mua hàng lần đầu để đưa ra ưu đãi đặc biệt trong tháng thứ hai, nhằm khuyến khích họ quay lại.
   - **Gửi email nhắc nhở**: Thực hiện chiến dịch email marketing để nhắc nhở khách hàng về sản phẩm họ đã xem hoặc mua trước đó.
   - **Chạy các chương trình khuyến mãi cho sản phẩm liên quan:**: Cung cấp các mã giảm giá hoặc ưu đãi đặc biệt cho các sản phẩm thường được mua cùng nhau trong thời gian giới hạn. Điều này có thể kích thích nhu cầu mua sắm.
   - **Xây dựng gói sản phẩm hoặc combo:**: Tạo ra các gói sản phẩm với các sản phẩm thường được mua cùng nhau với mức giá ưu đãi. Điều này không chỉ khuyến khích khách hàng mua thêm mà còn làm tăng giá trị đơn hàng trung bình.
   - **Gợi ý trong giỏ hàng:**
Khi khách hàng thêm một sản phẩm vào giỏ hàng, hiển thị một thông báo với các gợi ý sản phẩm liên quan, khuyến khích họ thêm các sản phẩm bổ sung vào giỏ hàng.

3. **Chú trọng đến nhóm khách hàng có nguy cơ rời bỏ**:
   - **Triển khai các chương trình tái kích hoạt**: Gửi email hoặc tin nhắn cá nhân với ưu đãi đặc biệt nhằm khuyến khích họ quay lại.
   - **Khảo sát ý kiến khách hàng**: Thực hiện khảo sát để tìm hiểu nguyên nhân khách hàng không quay lại, từ đó cải thiện sản phẩm và dịch vụ.

4. **Cải thiện chiến dịch marketing**:
   - **Sử dụng đa kênh tiếp thị**: Kết hợp các kênh truyền thông như mạng xã hội, email marketing và quảng cáo trực tuyến để tăng cường khả năng tiếp cận khách hàng.
   - **Chạy các chương trình ưu đãi trong thời điểm doanh thu thấp**: Đưa ra các khuyến mãi hấp dẫn trong các tháng doanh thu thấp để thu hút khách hàng và tăng cường doanh thu.

### **Dữ liệu ở thị trường Anh**

```python
df_uk = df1[df1['Country']=='United Kingdom']
df_uk.head()
```

```python
df_uk['InvoiceMonth'] = df_uk['InvoiceDate'].apply(get_month)
df_uk.head()
```

**Tính toán số lượng khách hàng ở mỗi tháng**

```python
#Tính toán số lượng khách hàng ở mỗi tháng
df_monthly_active = df_uk.groupby('InvoiceMonth')['CustomerID'].nunique().reset_index()
df_monthly_active.columns = ['InvoiceMonth', 'Number_of_customers']
df_monthly_active
```

```python
#Biểu đồ số lượng khách hàng theo tháng
plt.figure(figsize=(15, 5))
sns.barplot(data=df_monthly_active, x="InvoiceMonth", y="Number_of_customers")
plt.show()
```

**Xác định khách hàng cũ và khách hàng mới**

```python
#Tháng mua hàng đầu tiên của từng khách hàng
df_uk['FirstMonth'] = df_uk.groupby('CustomerID')['InvoiceMonth'].transform('min')
df_uk.head()
```

```python
#Xác định khách hàng cũ và khách hàng mới:
def check_type(row):
    if row["InvoiceMonth"] == row["FirstMonth"]:
        return "New"
    elif row["InvoiceMonth"] > row["FirstMonth"]:
        return "Existing"

df_uk["UserType"] = df_uk[["InvoiceMonth", "FirstMonth"]].apply(check_type, axis=1)
```

```python
df_uk.head()
```

**Số lượng khách hàng mới theo tháng**

```python
#Đếm số lượng khách hàng mới từng tháng
df_new_customer_uk = df_uk[df_uk["UserType"]=="New"].groupby("InvoiceMonth")["CustomerID"].nunique()
df_new_customer_uk
```

**Số lượng khách hàng đã có sẵn theo từng tháng**

```python
#Đếm số lượng khách hàng đã có sẵn theo từng tháng
df_existing_customer_uk = df_uk[df_uk["UserType"]=="Existing"].groupby("InvoiceMonth")["CustomerID"].nunique()
df_existing_customer_uk
```

**So sánh số lượng khách hàng mới và khách hàng cũ theo từng tháng trong năm (theo dạng tỷ lệ)**

```python
df_user_ratio_uk = df_new_customer_uk / df_existing_customer_uk
df_user_ratio_uk = df_user_ratio_uk.reset_index()
df_user_ratio_uk.columns = ["InvoiceMonth", "Ratio"]
df_user_ratio_uk
```

### **4.1. Tỷ lệ khách hàng quay lại (Retention Rate)**

Tài liệu tham khảo: https://blog.tomorrowmarketers.org/tang-ty-le-giu-chan-khach-hang-customer-retention-rate-the-nao/

Trong phân tích về tỷ lệ quay lại của khách hàng, chúng tôi sẽ sử dụng dữ liệu có các đơn hàng với giá trị Quantity > 0. Bởi vì chỉ những đơn hàng có quantity > 0 phản ánh một giao dịch mua bán thành công, nơi khách hàng đã thực sự sử dụng sản phẩm hoặc dịch vụ.

```python
df_uk.head()
```

```python
df_uk.describe()
```

```python
df2 = df_uk.copy()

#Lọc ra các giao dịch có số lượng lớn hơn 0
df2 = df2[df2['Quantity'] > 0]
df2.sample(10)
```

```python
#Thời gian khách hàng quay lại để mua hàng
df2['CohortIndex'] = (df2['InvoiceMonth'].dt.year - df2['FirstMonth'].dt.year)*12 + (df2['InvoiceMonth'].dt.month - df2['FirstMonth'].dt.month)
df2.head()
```

```python
#Calculate cohort

cohort_df = df2.groupby(['FirstMonth','InvoiceMonth', 'CohortIndex'])['CustomerID'].nunique().reset_index()
cohort_df = cohort_df.rename(columns={'CustomerID':'CustomerReturn'})
cohort_df.head()
```

```python
#Tạo hàm chuyển đổi datetime
def convert_date(df,col):
  return df[col].dt.strftime('%Y-%m-%d')

cohort_df['FirstMonth'] = convert_date(cohort_df,'FirstMonth')
cohort_df['InvoiceMonth'] = convert_date(cohort_df,'InvoiceMonth')
```

```python
cohort_df.head()
```

```python
#Tạo bảng "Tỷ lệ giữ chân khách hàng quay lại"
cohort_table = cohort_df.pivot_table(index='FirstMonth',
                                     columns='CohortIndex',
                                     values='CustomerReturn')

cohort_table.head()
```

```python
# Tính tổng số khách hàng trong mỗi nhóm (số khách hàng trong tháng đầu tiên)
cohort_sizes = cohort_table.iloc[:, 0]
```

```python
# Tính toán tỷ lệ giữ chân
retention_rate = cohort_table.divide(cohort_sizes, axis=0) * 100
retention_rate = retention_rate.round(2)
retention_rate.head()
```

```python

plt.figure(figsize=(15, 10))
ax = sns.heatmap(retention_rate, annot=True, fmt=".2f", cmap="Greens", cbar=True)
ax.xaxis.set_ticks_position('top')

for text in ax.texts:
    text.set_text(f'{text.get_text()}%')

plt.title('Retention rate')
plt.xlabel('Month')
plt.ylabel('First Month')
plt.show()
```

Dựa vào bảng cohort về tỷ lệ khách hàng quay lại mua hàng của cửa hàng, ta có thể đưa ra một số nhận xét sau:

- Tỷ lệ giữ chân khách hàng nhìn chung thấp hơn so với mặt bằng chung của ngành bán lẻ. Theo trang Sprinklr (https://www.sprinklr.com/blog/average-customer-retention-rate/#toc-2) tỷ lệ retention rate trung bình của ngành hàng bán lẻ trên các sàn thương mại điện tử là khoảng 38%. Ngoại trừ nhóm khách hàng tháng 12/2010, các nhóm khách hàng còn lại đều có tỷ lệ giữ chân sau 12 tháng thấp hơn mức trung bình của ngành bán lẻ.
Điều này cho thấy cửa hàng đang gặp khó khăn trong việc giữ chân khách hàng quay lại mua hàng sau lần đầu tiên.

- Xu hướng tỷ lệ giữ chân khách hàng giảm dần theo thời gian. Vì khách hàng chủ yếu của cửa hàng này chính là các đại lý bán lẻ. Việc tìm kiếm đối tác mới có thể khó khăn hơn vì các sản phẩm quà tặng thường có nhiều nhà cung cấp khác nhau. Điều này khiến cho sự cạnh tranh cao và khách hàng đại lý có thể dễ dàng chuyển sang các nhà cung cấp khác.
- Nhìn chung số lượng khách hàng mới có giao dịch ở tháng đầu tiên có dấu hiệu giảm dần theo thời gian. Những đại lý mới có xu hướng mua sắm thử một lần nhưng không duy trì tương tác lâu dài và còn phụ thuộc vào nhu cầu bán lẻ (tồn kho, sự phổ biến của sản phẩm) của họ và tình hình thị trường.

### **4.2. Phân khúc khách hàng - Phân tích mô hình RFM**

Tài liệu tham khảo:

1/ https://blog.tomorrowmarketers.org/phan-tich-rfm-la-gi/

2/ https://www.optimove.com/resources/learning-center/rfm-segmentation

Sử dụng mô hình RFM để phân tích và phân khúc khách hàng dựa trên hành vi mua hàng của họ.

```python
df2.head()
```

```python
df2.info()
```

```python
df2.isna().sum()
```

```python
#The Max Purchase Date of each customer
import datetime
act_date = max(df2['InvoiceDate'] + dt.timedelta(1))
```

```python
rfm = df2.groupby('CustomerID').agg({'InvoiceDate': lambda x: (act_date - x.max()).days ,
                                    'InvoiceNo': 'nunique',
                                    'Revenue': 'sum'})
```

```python
rfm.rename(columns = {'InvoiceDate' : 'Recency',
                      'InvoiceNo' : 'Frequency',
                      'TotalSum' : 'MonetaryValue'}, inplace = True)
rfm.head()
```

```python
rfm.describe()
```

```python
quantiles = rfm.quantile(q = [0.25, 0.50, 0.75])
quantiles
```

```python
def recency_scoring(rfm):
    if rfm.Recency <= 18.0:
        recency_score = 4
    elif rfm.Recency <= 51.0:
        recency_score = 3
    elif rfm.Recency <= 145:
        recency_score = 2
    else:
        recency_score = 1
    return recency_score

rfm['Recency_Score'] = rfm.apply(recency_scoring, axis=1)
rfm.sample(10)
```

```python
def frequency_scoring(rfm):
    if rfm.Frequency >= 4:
        frequency_score = 4
    elif rfm.Frequency >= 2:
        frequency_score = 3
    elif rfm.Frequency >= 1:
        frequency_score = 2
    else:
        frequency_score = 1
    return frequency_score

rfm['Frequency_Score'] = rfm.apply(frequency_scoring, axis=1)
rfm.sample(10)
```

```python
def monetary_scoring(rfm):
    if rfm.Revenue >= 1285.05:
        monetary_score = 4
    elif rfm.Revenue >= 547.60:
        monetary_score = 3
    elif rfm.Revenue >= 240.55:
        monetary_score = 2
    else:
        monetary_score = 1
    return monetary_score

rfm['Monetary_Score'] = rfm.apply(monetary_scoring, axis=1)
rfm.sample(10)
```

```python
#Tính điểm RFM cho mỗi khách hàng
def rfm_scoring(customer):
    return str(int(customer['Recency_Score'])) + str(int(customer['Frequency_Score'])) + str(int(customer['Monetary_Score']))


rfm['Customer_RFM_Score'] = rfm.apply(rfm_scoring, axis=1)
rfm.sample(10)
```

```python
#Tính điểm RFM
rfm['RFM_Score'] = rfm['Recency_Score'] + rfm['Frequency_Score'] + rfm['Monetary_Score']
rfm.sample(10)
```

```python
rfm['RFM_Score'].unique()
```

```python
rfm.reset_index(inplace=True)
rfm.sample(10)
```

```python
def assign_label(row):
    r, f, m = row['Recency_Score'], row['Frequency_Score'], row['Monetary_Score']

    if r == 4 and f == 4 and m == 4:
        return 'Champion'
    elif r == 3 and f in (3, 4) and m in (3, 4):
        return 'Loyal Customer'
    elif r == 4 and f in (1, 2, 3, 4) and m in (3, 4):
        return 'Potential Loyalist'
    elif r == 4 and f == 1 and m == 1:
        return 'Recent Customers'
    elif r == 3 and f == 2 and m == 2:
        return 'Promising'
    elif r == 3 and f == 1 and m == 1:
        return 'Customer Needing Attention'
    elif r == 2 and f == 1 and m == 1:
        return 'About to sleep'
    elif r == 1 and f == 2 and m == 2:
        return 'At risk'
    elif r == 1 and f == 3 and m == 3:
        return 'Can\'t lose them'
    elif r == 2 and f == 1 and m == 1:
        return 'Hibernating'
    elif r == 1 and f == 1 and m == 1:
        return 'Lost'
    else:
        return 'Normal'
rfm['Customer_Segment'] = rfm.apply(assign_label, axis=1)
rfm.sample(10)
```

Dựa vào các tiêu chí tương đồng về hành vi và giá trị khách hàng. Chúng ta có thể nhóm các phân khúc thành một nhóm tương đồng với nhau để đưa ra các chiến lược phát triển tiếp cận dễ dàng hơn.

**1. Nhóm khách hàng trung thành (Loyal Customers):**

Champion: Đây là nhóm khách hàng tốt nhất, thường xuyên mua và có giá trị cao.

Loyal Customer: Khách hàng trung thành, mua thường xuyên, nhưng có thể chưa đạt đến cấp độ cao như Champion.

Can't lose them: Khách hàng trung thành có nguy cơ không quay lại, cần giữ chân.


**2. Nhóm khách hàng tiềm năng hoặc đang phát triển (Potential Customers):**

Potential Loyalist: Khách hàng có tiềm năng trở thành trung thành trong tương lai.

Promising: Khách hàng hứa hẹn, có thể chưa mua nhiều nhưng có dấu hiệu tích cực.

Recent Customers: Khách hàng mới, có thể phát triển thành khách hàng trung thành.

**3. Nhóm khách hàng có rủi ro (Risk customers)**

At risk: Khách hàng đang có nguy cơ không quay lại, cần hành động để giữ chân.

Customer Needing Attention: Khách hàng cần quan tâm thêm, có thể vì họ đã bắt đầu giảm mức độ mua hàng.


**4. Nhóm khách hàng không hoạt động hoặc đã bỏ rơi (Churned Customers)**

Hibernating: Khách hàng không còn mua hàng hoặc rất ít mua, đã dừng hoạt động.

About to sleep: Khách hàng sắp "ngủ quên", không còn hoạt động tích cực.

**5. Nhóm khách hàng bình thường (Normal CUstomers)**

Normal: Khách hàng bình thường, không có đặc điểm nổi bật, mua với tần suất trung bình.


```python
#Nhóm các phân khúc lại thành nhóm mới
def assign_group(segment):
    if segment in ['Champion', 'Loyal Customer', 'Can\'t lose them']:
        return 'Loyal Customers'
    elif segment in ['Potential Loyalist', 'Promising', 'Recent Customers']:
        return 'Potential Customers'
    elif segment in ['At risk', 'Customer Needing Attention']:
        return 'Risk Customers'
    elif segment in ['About to sleep', 'Hibernating', 'Lost']:
        return 'Churned Customers'
    else:
        return 'Normal Customers'

# Thêm cột Group_Segment
rfm['Group_Segment'] = rfm['Customer_Segment'].apply(assign_group)


rfm.sample(10)
```

      - Dịch vụ gói quà miễn phí: Đối với khách hàng trong nhóm nguy cơ, cung cấp dịch vụ gói quà miễn phí như một ưu đãi kèm theo. Điều này không chỉ gia tăng sự tiện lợi mà còn khuyến khích họ quay lại mua hàng.

      - Chương trình thành viên quà tặng nơi khách hàng có thể tích lũy điểm cho mỗi lần mua hàng và đổi điểm lấy các sản phẩm quà tặng độc quyền. Chương trình này không chỉ thu hút mà còn tăng tần suất mua hàng.

### **4.3. Phân tích giỏ hàng của khách hàng (Basket Analysis)**

Trong phần phân tích này, chúng ta sử dụng Apriori Algorithm là một thuật toán rất phổ biến trong việc phân tích giỏ hàng (basket analysis), đặc biệt trong ngành bán lẻ để tìm ra những nhóm sản phẩm mà khi khách hàng mua một sản phẩm thì họ có xu hướng mua thêm những sản phẩm khác.

Lợi ích của việc áp dụng Apriori Algorithm:
- Tăng doanh thu: Đề xuất các sản phẩm liên quan giúp tăng giá trị đơn hàng.
- Hiểu hành vi khách hàng: Phát hiện các mẫu mua hàng giúp chúng ta hiểu hơn về nhu cầu và hành vi của khách hàng.
- Tối ưu chiến lược bán hàng: Giúp lên kế hoạch khuyến mãi, bày trí sản phẩm một cách hiệu quả hơn.

```python
# Tạo giỏ hàng (basket) theo InvoiceNo và StockCode
basket = customer_segment.groupby(['InvoiceNo', 'StockCode'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')

# Chuyển đổi thành giá trị bool để dễ xử lý
basket = basket.astype(bool)

# Tạo từ điển để đếm số lần xuất hiện của các cặp sản phẩm
pair_occurrence = {}

for idx, row in basket.iterrows():
    # Lấy danh sách các sản phẩm có trong giỏ hàng
    items = row[row].index.tolist()

    # Tạo các cặp sản phẩm
    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            sorted_pair = tuple(sorted([items[i], items[j]]))  # Sắp xếp cặp để tránh trùng lặp
            if sorted_pair in pair_occurrence:
                pair_occurrence[sorted_pair] += 1
            else:
                pair_occurrence[sorted_pair] = 1

pair_df = pd.DataFrame(list(pair_occurrence.items()), columns=['ProductPair', 'Count'])
pair_df_sorted = pair_df.sort_values(by='Count', ascending=False).head(10)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Áp dụng thuật toán Apriori để tìm các sản phẩm thường xuất hiện cùng nhau
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)

# Tạo quy tắc kết hợp
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Lọc ra các cặp sản phẩm mà có tỉ lệ mua cùng nhau cao
filtered_rules = rules[(rules['lift'] > 1) & (rules['confidence'] > 0.5)]

# Chuyển các cặp sản phẩm thành sets để loại bỏ các cặp trùng lặp như (A, B) và (B, A)
filtered_rules['pair'] = filtered_rules.apply(lambda x: frozenset(list(x['antecedents']) + list(x['consequents'])), axis=1)

# Loại bỏ các cặp trùng lặp
unique_pairs = filtered_rules.drop_duplicates(subset='pair')

```

```python
def get_samples_by_stockcodes(df, stock_codes, n=1):
    samples = []
    for code in stock_codes:
        sample = df[df['StockCode'] == code].sample(n=n)
        sample['StockCode'] = code
        samples.append(sample)
    return pd.concat(samples, ignore_index=True)

# Sử dụng hàm
stock_codes = ['20727', '22384', '20725', '22384', '22386', '85099B']
sampled_data = get_sample_by_stockcodes(df3, stock_codes, n=1)
```
